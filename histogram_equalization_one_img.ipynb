```python
# Compute histograms for each color channel separately
hist_red, bins_red = np.histogram(img[:, :, 0].flatten(), 256, [0, 256])
hist_green, bins_green = np.histogram(img[:, :, 1].flatten(), 256, [0, 256])
hist_blue, bins_blue = np.histogram(img[:, :, 2].flatten(), 256, [0, 256])

# Compute cumulative distribution functions (CDFs) for each channel
cdf_red = hist_red.cumsum()
cdf_green = hist_green.cumsum()
cdf_blue = hist_blue.cumsum()

# Normalize the CDFs
cdf_red_normalized = cdf_red * float(hist_red.max()) / cdf_red.max()
cdf_green_normalized = cdf_green * float(hist_green.max()) / cdf_green.max()
cdf_blue_normalized = cdf_blue * float(hist_blue.max()) / cdf_blue.max()

# Plot CDFs
plt.plot(cdf_red_normalized, color='r', label='CDF Red')
plt.plot(cdf_green_normalized, color='g', label='CDF Green')
plt.plot(cdf_blue_normalized, color='b', label='CDF Blue')

# Plot histograms
plt.hist(img[:, :, 0].flatten(), 256, [0, 256], color='r', alpha=0.5, label='Histogram Red')
plt.hist(img[:, :, 1].flatten(), 256, [0, 256], color='g', alpha=0.5, label='Histogram Green')
plt.hist(img[:, :, 2].flatten(), 256, [0, 256], color='b', alpha=0.5, label='Histogram Blue')

plt.xlim([0, 256])
plt.legend(loc='upper left')
plt.show()
```

```python
equ_r = cv.equalizeHist(img[:, :, 0])
equ_b = cv.equalizeHist(img[:, :, 1])
equ_g = cv.equalizeHist(img[:, :, 2])
img_out = cv.merge((equ_b, equ_g, equ_r))
```

```python
cv.imshow("image histogram equalized",img_out)
```
